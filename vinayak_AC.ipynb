```
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Create age bins
age_bins = [0, 20, 30, 40, 50, 60]
age_labels = ['0-20', '21-30', '31-40', '41-50', '51-60']
df['age_group'] = pd.cut(df['housing_median_age'], bins=age_bins, labels=age_labels)

# Group by age group and calculate average house value
grouped = df.groupby('age_group')['median_house_value'].mean()

# Plotting
plt.figure(figsize=(8, 5))
grouped.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Median House Value by Housing Age Group')
plt.xlabel('Housing Age Group')
plt.ylabel('Median House Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Plotting the histogram
plt.figure(figsize=(8, 5))
plt.hist(df['median_house_value'], bins=30, color='teal', edgecolor='black')
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Create the boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(df['median_house_value'], vert=False, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='navy'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='gray'),
            capprops=dict(color='gray'),
            flierprops=dict(marker='o', color='orange', alpha=0.5))

plt.title('Boxplot of Median House Value')
plt.xlabel('Median House Value')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Plotting the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['median_income'], df['median_house_value'], alpha=0.5, color='purple', edgecolors='w')
plt.title('Median Income vs Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Plotting the density chart
plt.figure(figsize=(10, 6))
sns.kdeplot(df['median_house_value'], shade=True, color='skyblue', linewidth=2)
plt.title('Density Plot of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.show()
```
